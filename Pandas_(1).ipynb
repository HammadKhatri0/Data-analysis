```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as pltp
```

```python
df = pd.read_csv("C:/Users/Hammad/Desktop/Pakistan Largest Ecommerce Dataset (1).csv", dtype="unicode")
```

```python
status = df['status'].value_counts().plot(kind = 'bar', figsize = (10, 5))
for container in status.containers:
    status.bar_label(container)
pltp.title("Order Status Frequency")
pltp.xlabel("Order Status")
pltp.xticks(rotation=45)
pltp.ylabel("Frequency in percentage")
pltp.legend()
pltp.show()
pltp.close()
print("\u2022"+"\t"+"""According to data on large masses buyers are cancelling their orders which shows uncertainity in e- commerce market due
        to insufficient checking tools of product and seller.""")
```

```python
dt = df['payment_method'].value_counts(normalize=True)
labels = df['payment_method'].unique().tolist()
del labels[-1]
labels[1],labels[4]=labels[4],labels[1]
pltp.pie(dt, labels=labels, radius = 1.5, autopct='%1.1f%%')
pltp.title("Payment Method Frequency", pad=50)
pltp.legend(bbox_to_anchor=(2,1.5))
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to data, on large masses buyers are using COD and few are also using online transactions.")
```

```python
freq_cat = df['category_name_1'].value_counts(normalize=True).mul(100).plot(kind = 'bar', figsize = (10, 5))
for container in freq_cat.containers:
    freq_cat.bar_label(container, fmt='%.2f')
pltp.title("Frequency of Categories")
pltp.xlabel("Categories")
pltp.xticks(rotation=45)
pltp.ylabel("Frequency in percentage")
pltp.legend()
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to data most buyers are buying from the Mobiles & Tablets Category.")
```

```python
df['grand_total']=df['grand_total'].astype('float')
df['Date']=pd.to_datetime(df.Date, errors='coerce').dt.strftime('%Y-%m')
gk = df.groupby("Date")["grand_total"].sum()/10**6
sales_mon = gk.plot(kind = "bar", figsize = (15, 5))
for container in sales_mon.containers:
    sales_mon.bar_label(container, fmt='%.1f')
pltp.title("Monthly Sales")
pltp.xlabel("Date")
pltp.xticks(rotation=45)
pltp.ylabel("Sales in million")
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to data in November 2017 large amount of sales have been recorded.")
```

```python
yearly_sales = df.groupby("Year")["grand_total"].sum()/10**6
labels = df['Year'].unique().tolist()
del labels[-1]
pltp.pie(yearly_sales, labels = labels, autopct='%1.1f%%')
pltp.title("Payment Method Frequency")
pltp.legend(bbox_to_anchor=(1,1))
pltp.title("Yearly Sales")
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to data in 2017 highest sales have been recorded.")
```

```python
df['discount_amount']=df['discount_amount'].astype('float')
dk = df.groupby("category_name_1")["discount_amount"].sum()/10**5
disc_cat = dk.plot(kind = "bar", figsize = (10, 5))
for container in disc_cat.containers:
    disc_cat.bar_label(container, fmt='%.0f')
pltp.title("Discount on Categories")
pltp.xlabel("Categories")
pltp.xticks(rotation=45)
pltp.ylabel("Discount amount in million")
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to data Mobiles and Tablets gives the large discount to customers.")
```

```python
df["qty_ordered"] = df["qty_ordered"].astype('float')
sales_best = df.groupby("category_name_1")["qty_ordered"].sum().plot(kind = 'bar', figsize = (10, 5))
for container in sales_best.containers:
    sales_best.bar_label(container, fmt='%.0f')
pltp.title("Best Selling Category")
pltp.xlabel("Categories")
pltp.xticks(rotation=55)
pltp.ylabel("Quantity ordered")
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to the data the best-selling category is Mobiles and Tables.")
```

```python
fraud = df.set_index('status').filter(like='fraud', axis=0)
f_data = fraud["payment_method"].value_counts(normalize=True).sort_values(ascending=True).mul(100)
f_labels = fraud['payment_method'].unique().tolist()
pltp.pie(f_data, labels = f_labels, autopct='%1.1f%%')
pltp.title("Fraud by categories")
pltp.legend(bbox_to_anchor=(1,1))
pltp.show()
pltp.close()
print("\u2022"+"\t"+"According to the data 60% fraud is commited by Payaxis payment method.")
```
